```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = r"C:\Users\sinbek\OneDrive - Aker BP\Documents\AI Drill\Time-series-stuck-pipe-prediction\TSPP\TSPP\TSPP1-15s.csv"
data = pd.read_csv(file_path)

# Parse the date column
data['date'] = pd.to_datetime(data['date'])

# Create subplots for each parameter
fig, axes = plt.subplots(nrows=len(data.columns) - 1, ncols=1, figsize=(10, 20), sharey=True)

# Plot each parameter in a separate subplot
for i, column in enumerate(data.columns[1:]):  # Skip the date column
    axes[i].plot(data[column], data['date'])
    axes[i].set_xlabel(column)
    axes[i].invert_yaxis()  # Invert the y-axis to have the oldest data at the top
    axes[i].grid(True)

# Set the y-axis label for all subplots
axes[0].set_ylabel('Date')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
